**Reading Images**

Number of ways to read an image:

```python
#Using Pillow library
# pillow is used for basic manipulation of image like reading, loading, resizing images
from PIL import Image
from matplotlib import pyplot as plt
img = Image.open("/content/ak.jpg")
print(type(img))
# Check type of Image
plt.imshow(img)
```

```python
from skimage import io
image = io.imread("/content/ak.jpg")
print(type(image))
# Check type of Image
plt.imshow(image)
```

```python
#cv2 is programmiing function mainly aim at computer vision
import cv2
img = cv2.imread("/content/ak.jpg")
print(type(img))
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```
