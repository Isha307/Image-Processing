<H3> Adding one image over another</H3>

```python
img1 = Image.open("/content/c.jpg")
img2 = Image.open("/content/ak.jpg")
img1_copy = img1.copy()
img1_copy.paste(img2,(50,50))
plt.imshow(img1_copy)
```

<h3> Rotate Image</h3>

```python
img = Image.open("/content/ak.jpg")
img90 = img.rotate(90)
plt.imshow(img90)
```

```python
img_flipLR = img.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(img_flipLR)
```

```python
img_flipTB = img.transpose(Image.FLIP_TOP_BOTTOM)
plt.imshow(img_flipTB)
```
