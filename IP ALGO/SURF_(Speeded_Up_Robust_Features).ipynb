Basically used for image comaprison and invarint to image brightness and contrast.

Descriptor define the feature no matter what rotation, lightning, orientation is.

## Hessian Matrix
A Hessian matrix is a square matrix whose elements are second-order partial derivatives of a given function.

Determinants can be used to classify critical points of differentiate functions. For example, if f : **ℝ2 → ℝ** is a function with continuous second partial derivatives fxx, fxy, fyx, and fyy, then the matrix

**MatrixForm** [Hf = {{fxx, fxy}, {fyx, fyy}}]

is a Hessian matrix. Its determinant Det[Hf] is called the discriminant of f.

```python
sift = cv2.xfeatures2d.SIFT_create()
surf = cv2.xfeatures2d.SURF_create()
orb = cv2.ORB_create(nfeatures=1500)
```

We find the keypoints and descriptors of each spefic algorythm.

A keypoint is the position where the feature has been detected, while the descriptor is an array containing numbers to describe that feature.

```python
keypoints_sift, descriptors = sift.detectAndCompute(img, None)
keypoints_surf, descriptors = surf.detectAndCompute(img, None)
keypoints_orb, descriptors = orb.detectAndCompute(img, None)
```
