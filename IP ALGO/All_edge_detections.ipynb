```python
from skimage.filters import  roberts, prewitt, sobel, scharr, laplace
import cv2
```

```python
img = cv2.imread('/content/img.jpeg')
```

```python
img.shape
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
edge1 = roberts(img)
edge2 = sobel(img)
edge3 = prewitt(img)
```

```python
figure = plt.figure(figsize=(12, 12))
subplot1 = figure.add_subplot(1, 3, 1)
subplot1.imshow(edge1)
subplot2 = figure.add_subplot(1, 3, 2)
subplot2.imshow(edge2)
subplot3 = figure.add_subplot(1, 3, 3)
subplot3.imshow(edge3)
```

```python
 edge5 = cv2.Canny(img,100,200)
 plt.imshow(edge5)
```

```python
edge6 = cv2.Laplacian(img, cv2.CV_64F)
plt.imshow(edge6)
```

```python
blur_img = cv2.GaussianBlur(img, (3, 3), 0)
edge7 = cv2.Laplacian(blur_img, cv2.CV_64F)
plt.imshow(edge7)
```
